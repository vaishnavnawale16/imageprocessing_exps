```python
# Importing the modules
import cv2
import numpy as np
# Importing the necessary function for image display in Colab
from google.colab.patches import cv2_imshow

# Reading the image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/download.jpg')

# Applying the filter
medianBlur = cv2.medianBlur(image, 9)

# Showing the image using cv2_imshow for Colab compatibility
cv2_imshow(image)  # Displaying the original image
cv2_imshow(medianBlur) # Displaying the blurred image

cv2.waitKey()
cv2.destroyAllWindows()
```
