```python
# Importing the modules
import cv2
import numpy as np
# Import the necessary patch for cv2_imshow
from google.colab.patches import cv2_imshow

# Reading the image
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/download.jpg')

# Applying the filter
averageBlur = cv2.blur(image, (5, 5))

# Showing the image
# Use cv2_imshow instead of cv2.imshow
cv2_imshow(image)
cv2_imshow(averageBlur)

cv2.waitKey()
cv2.destroyAllWindows()
```
